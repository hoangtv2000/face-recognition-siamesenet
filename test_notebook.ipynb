```python
import glob
import pandas as pd

result = pd.DataFrame(columns=['image', 'label'])

for full_img_dir in glob.glob('processed_data\processed_face_aligned\*\*'):
    label_ = full_img_dir.split('\\')[-2]
    dict_ = {'image': full_img_dir, 'label': label_}
    dict_ = pd.DataFrame([dict_])
    result = pd.concat([result, dict_])

result.to_csv('processed_data\processed_face_aligned.csv')
```

```python
import glob
import pandas as pd
import numpy as np
from tensorflow.keras.utils import to_categorical

NUM_CLASSES = 1002

def convert2categorical(labels):
    categories = []
    for index, row in labels.iterrows():
        categories.append(to_categorical(row[2], num_classes = NUM_CLASSES+1))
    return np.array(categories)
    
labels = pd.read_csv('processed_data\embedded_face\embs_class_20180402-114759.csv')
labels.columns = ['index', 'emb', 'class']
print(labels)
categories = convert2categorical(labels)

labels_train = labels.sample(frac = 0.8, random_state = 0)
labels_val = labels.drop(labels_train.index)
categories_train = convert2categorical(labels_train)
categories_val = convert2categorical(labels_val)

labels_train.reset_index(inplace = True)
labels_val.reset_index(inplace = True)

print(labels_train.head)
print(labels_val.head)
print(categories_train.shape)
print(categories_val.shape)
```

```python
import pandas as pd
import numpy as np
import os

predictions = pd.read_csv('predictions\output_facenet_adacos_20180402-114759.csv')
NUM_CLASSES = len(os.listdir('processed_data\\embedded_face\\20180402-114759\\processed_face_aligned'))

overall_preds = []

cnt = 0
for i in range(predictions.shape[0]):
    if '_flip' in predictions['image'][i]:
        continue
    # print("{}/{}".format(i, labels.shape[0]))
    avg_probs = np.array([(predictions["class_{}".format(x)][i]) for x in range(1, NUM_CLASSES+1)])
    predicted_class = np.argmax(avg_probs) + 1
    
    print(predictions['image'][i].split('/')[4], '//', predicted_class)
    if predictions['image'][i].split('/')[4] == predicted_class:
        cnt+=1
```
